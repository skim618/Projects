```python
import pandas as pd 
import numpy as np
import folium
import openrouteservice as ors
from folium.plugins import FloatImage
import warnings
warnings.filterwarnings("ignore", category = UserWarning)
ORSkey = '5b3ce3597851110001cf62488ddcf95b7c114a33933d6481636caf73'

# read in data
weekdayRoutes = pd.read_csv('weekdayRoutes.txt', converters = {'Stores' : eval})
saturdayRoutes = pd.read_csv('saturdayRoutes.txt', converters = {'Stores' : eval})
northClosedWeekdayRoutes = pd.read_csv('northClosedWeekdayRoutes.txt', converters = {'Stores' : eval})
northClosedSaturdayRoutes = pd.read_csv('northClosedSaturdayRoutes.txt', converters = {'Stores' : eval})
locations = pd.read_csv('WarehouseLocations.csv')
storeNames = list(locations['Store'])

# extract Long/Lat data of warehouse locations
coords = locations[['Long','Lat']]
coords = coords.to_numpy().tolist()

# boot up client to OpenRouteServices
client = ors.Client(key = ORSkey)

# create list of colours
list_color = [
    '#FF6800',
    '#C0C0C0',
    '#808080',
    '#000000',
    '#FF0000',
    '#800000',
    '#FFFF00',
    '#808000',
    '#00FF00',
    '#008000',
    '#00FFFF',
    '#008080',
    '#0000FF',
    '#000080',
    '#FF00FF',
    '#800080',
    '#34495E'
]
```

```python
# create map
m = folium.Map(location = list(reversed(coords[2])), zoom_start=10)

index = 0
# visualise routes for weekdays
for route in weekdayRoutes['Stores']:
    feature_group = folium.FeatureGroup(str(route))
    for i in range(len(route)-1):
        ORSroute = client.directions(
            coordinates = [coords[storeNames.index(route[i])],coords[storeNames.index(route[i+1])]],
            profile = 'driving-hgv',
            format = 'geojson',
            validate = False
        )
        
        folium.PolyLine(locations = [list(reversed(coord))
                                        for coord in 
                                        ORSroute['features'][0]['geometry']['coordinates']], color=list_color[index]).add_to(feature_group)
        feature_group.add_to(m)
    index += 1
folium.LayerControl().add_to(m)
m
```
